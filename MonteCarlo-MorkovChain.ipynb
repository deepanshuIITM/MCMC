<font size="5"> Monte Carlo Simulation: Example 1

 Evaluate the integral:
 $$\int_0^1 cos(2\pi\sqrt{(1-x^{2}})dx$$

 Check whether the scaling of the intervals goes down as $\frac{1}{\sqrt{N}}$


<br> **Method 1:**
<br> $ \int_{0}^{1} cos(2\pi\sqrt{1- x^{2}} ) dx $
<br> x is chosen as Uniformly Distributed random variable: x ~ unif$[0, 1]$ and value of $cos(2\pi\sqrt{1- x^{2}})$ is being calculated. The mean of values of $cos(2\pi\sqrt{1- x^{2}})$ will give the integral value $ \int_{0}^{1} cos(2\pi\sqrt{1- x^{2}} ) dx $

```python
import random as rad
import numpy as np
import math as ma
import statistics as stats
import matplotlib.pyplot as plt
N_sim = 100
mean1 = np.zeros(N_sim)
mean2 = np.zeros(N_sim)
mean3 = np.zeros(N_sim)
mean4 = np.zeros(N_sim)
for sim in range (0,N_sim,1):
    N = [10,100,1000,10000]
    r = len(N)
    for k in range(0,r,1):
        n = N[k]
        f = 0
        for i in range(0,n,1):
            x = rad.uniform(0,1)
            f = f + ma.cos(2*ma.pi*ma.sqrt(1-x*x))
        if k ==0:
            mean1[sim] = f/n
        if k ==1:
            mean2[sim] = f/n
        if k==2:
            mean3[sim] = f/n
        if k==3:
            mean4[sim] = f/n
t1 = 1*np.ones(100)
t2 = 2*np.ones(100)
t3 = 3*np.ones(100)
t4 = 4*np.ones(100)
std = [stats.stdev(mean1),stats.stdev(mean2),stats.stdev(mean3),stats.stdev(mean4)]
print('Value of integral Using Monte Carlo Method =', stats.mean(mean4))
print('SD for n=10,100,1000,10000=',std)
```

```python
plt.rcParams["figure.figsize"] = [6,7];
plt.plot(t1, mean1, 'o',linewidth=2, markersize=1.5)
plt.plot(t2, mean2, 'o',linewidth=2, markersize=1.5)
plt.plot(t3, mean3, 'o',linewidth=2, markersize=1.5)
plt.plot(t4, mean4, 'o',linewidth=2, markersize=1.5)
plt.axhline(y = 0.35169394078, color='k',linewidth=1, linestyle='--',label='Actual')

plt.title('Variation of integral value for different N')
plt.xlabel('log(N)-------->')
plt.ylabel('Integral distribution-------->')
plt.grid()
```

<br> **Value of integral Using Monte Carlo Method = 0.35235209**
<br> Standard deviation is also being calculated for different sample size, it can be concluded that scaling of the interval goes down as $\frac{1}{\sqrt{N}}$:
<br> <br> $
\begin{matrix}
N & Standard Deviation (\sigma) \\

\end{matrix}
$

<br>  $\dfrac{\sigma_{1}}{\sigma_{2}}$ is in order of $\sqrt{\dfrac{N_{2}}{N_{1}}}$

<br> **Method 2:**
<br> As $x = [0, 1]$ and $y = [-1, 1]$, so x ~ unif(0,1) and y ~ unif(-1,1) is being chosen as two random variables and points were dumped into the rectangle of area 2 sq unit. Now at (x,y) function values is being calculated at sample size (N) and counted the numbers(n) when:
<br> 1. y $\leq$ function value if y > 0, and
<br> 2. y $\geq$ function value if y < 0
<br> so area under the curve = $\frac{n}{N}$ (Area of Rectangle)
<br> $A_{curve}$ = 2$\frac{n}{N}$

```python
a1 = np.array([]);b1 = np.array([]); a2 = np.array([]);b2 = np.array([]);
x = np.linspace(0,1,1000);
y = np.cos(2*ma.pi*((1-x**2)**0.5))
plt.plot(x,y); plt.grid()
plt.axhline(y = 0, color='k',linewidth=1, linestyle='--')
plt.axhline(y = 1, color='r',linewidth=2, linestyle='-')
plt.axhline(y = -1, color='r',linewidth=2, linestyle='-')
plt.axvline(x = 0, color='r',linewidth=2, linestyle='-')
plt.axvline(x = 1, color='r',linewidth=2, linestyle='-')
for i in range(0,10000,1):
    x1 = rad.uniform(0,1)
    y1 = rad.uniform(-1,1)
    fun = np.cos(2*ma.pi*((1-x1**2)**0.5))
    if 0 <= y1 <= fun:
        a1 = np.append(a1, x1)
        b1 = np.append(b1, y1)
    if 0 >= y1 >= fun:
        a2 = np.append(a2, x1)
        b2 = np.append(b2, y1)
plt.plot(a1, b1, 'o',linewidth=2, markersize=1.5)
plt.plot(a2, b2, 'o',linewidth=2, markersize=1.5)
plt.title('Monte Carlo Simulation for Integration')
plt.xlabel('x-------->');plt.ylabel('f(x)-------->');
```

<font size="5"> Monte Carlo Simulation: Example 2

Let the radius of the top plate is $r_{p}$, radius of the bottom plate is $r_{b}$ and radius of bolt is R, if the centre offset between top plate and bottom plate is r then:

<br> **CASE I: r $\geq |r_{p}  r_{b}|$**
<br> If the offset between the centre of two plate is more than their radius difference/ radial clearance, then the radius of the bolt which can be fitted into, will be:
<br> $R \leq \dfrac{r_{p}  r + r_{b}}{2}$

<br> **CASE II: r $\leq |r_{p}  r_{b}|$**
<br> If the offset between the centre of two plate is less than their radius difference/ radial clearance, then the radius of the bolt which can be fitted into, will be:
<br> $R \leq min (r_{p}, r_{b}$)
<br> Also r = $\sqrt{(X_{p}-X_{b})^2 + (Y_{p}-Y_{b})^2}$, where
<br> $X_{p}$ is top plate x- location
<br> $Y_{p}$ is top plate y- location
<br> $X_{b}$ is bottom plate x- location
<br> $Y_{b}$ is bottom plate y- location

```python
import random as rad
import numpy as np
import math as ma
import statistics as stats
N = 100000
r = np.zeros(N)
i =0;
for k in range (0,N,1):
    Rb=np.random.normal(25.25,0.05/3)
    Rp=np.random.normal(25.15,0.05/3)
    R_bolt = np.random.normal(24.95,0.21/3)
    Tp_X = np.random.normal(100,0.2/3)
    Tp_Y = np.random.normal(100,0.2/3)
    Bp_X = np.random.normal(100,0.2/3)
    Bp_Y = np.random.normal(100,0.2/3)
    r[k] = ma.sqrt((Tp_X-Bp_X)**2 + (Tp_Y-Bp_Y)**2)
    if r[k] <= abs(Rb-Rp):
        if R_bolt <= Rp:
            i = i+1
    if r[k] >=abs(Rb-Rp):
        if R_bolt <= (Rp+Rb-r[k])/2:
            i = i+1
print((1-i/N)*100,'Percentage of the time Bolt will not fit into the assembly')
print('Nominal Value of center offset of holes=', stats.mean(r))
print('Standard deviation of center offset=',stats.stdev(r))
```

<font size="5"> MCMC: Example 1

```python
import random as rad
import numpy as np
import math as ma
import statistics as stats
import matplotlib.pyplot as plt
N = 1000000; state = np.ones(N);p1 = 1/3;p2 = 2/3; a =0;b =0;c =0;d =0
for k in range(0,N-3,1):
    x = rad.uniform(0,1)
    if state[k]==1:
        if x < p2:
            state[k+1]=1;
        else:
            state[k+1]=2;
    if state[k] ==2:
        if x< p1:
            state[k+1] =2
        if p1<x<p2:
            state[k+1] =1
        else:
            state[k+1] =3
    if state[k] ==3:
        if x< p1:
            state[k+1] =3
        if p1<x<p2:
            state[k+1] =4
        else:
            state[k+1] =2
    if state[k]==4:
        if x < p2:
            state[k+1]=4;
        else:
            state[k+1]=3;
pi1 =np.zeros(N);pi2 =np.zeros(N);pi3 =np.zeros(N);pi4 =np.zeros(N);
for i in range (0,N,1):
    if state[i] ==1:
        a=a+1
        pi1[i]=a/(1+i)
    else:
        pi1[i]=a/(1+i)
    if state[i] ==2:
        b=b+1
        pi2[i]=b/(1+i)
    else:
        pi2[i]=b/(1+i)
    if state[i] ==3:
        c=c+1
        pi3[i]=c/(1+i)
    else:
        pi3[i]=c/(1+i)
    if state[i] ==4:
        d=d+1
        pi4[i]=d/(1+i)
    else:
        pi4[i]=a/(1+i)
print('Steady state probability for state 1 =',a/N)
print('Steady state probability for state 2 =',b/N)
print('Steady state probability for state 3 =',c/N)
print('Steady state probability for state 4 =',d/N)
#print(a+b+c+d)
```

```python
t = np.linspace(1,N,N)
plt.figure(figsize=(12,10))
plt.subplot(2,2,1); plt.plot(t,pi1,linewidth=1.5)
plt.xlabel('N-------->'); plt.ylabel('pi1-------->'),plt.grid()
plt.subplot(2,2,2); plt.plot(t,pi2,linewidth=1.5)
plt.xlabel('N-------->'); plt.ylabel('pi2-------->'),plt.grid()
plt.subplot(2,2,3); plt.plot(t,pi3,linewidth=1.5)
plt.xlabel('N-------->'); plt.ylabel('pi3-------->'),plt.grid()
plt.subplot(2,2,4); plt.plot(t,pi4,linewidth=1.5)
plt.xlabel('N-------->'); plt.ylabel('pi4-------->'),plt.grid()
plt.suptitle('Steady State Transition Probabilities')
```

Given Probability Transition matrix P:
<br> $$
\begin{pmatrix}
1-p & p & 0 & 0 \\
p & 1-2p & p & 0 \\
0 & p & 1-2p & p \\
0 & 0 & p & 1-p \\
\end{pmatrix}
$$
<br> For p = $\frac{1}{3}$ Probability Transition matrix P:
<br>$$
\begin{pmatrix}
2/3 & 1/3 & 0 & 0 \\
1/3 & 1/3 & 1/3 & 0 \\
0 & 1/3 & 1/3 & 1/3 \\
0 & 0 & 1/3 & 2/3 \\
\end{pmatrix}
$$
<br> Now, $P_{4\times4} \pi_{4\times1} = \pi_{4\times1}$ ..... (1)
<br> Where $\pi_{4\times1} = [\pi_{1} , \pi_{2}, \pi_{3}, \pi_{4}]$
<br> From (1) we will get the following equations:
<br> $$ \frac{2}{3} \pi_{1} + \frac{1}{3} \pi_{2} = \pi_{1} ........(i)$$
<br> $$ \frac{1}{3} \pi_{1} + \frac{1}{3} \pi_{2} + \frac{1}{3} \pi_{3} = \pi_{2} .........(ii)$$
<br> $$ \frac{1}{3} \pi_{2} + \frac{1}{3} \pi_{3} + \frac{1}{3} \pi_{4} = \pi_{3} .........(iii)$$
<br> $$ \frac{1}{3} \pi_{3} + \frac{2}{3} \pi_{4} = \pi_{4} ........(iv)$$
<br> From equation (i) : $\pi_{1} = \pi_{2}$
<br> From equation (ii) : $\pi_{3} = \pi_{2}$  (Using  $\pi_{1} = \pi_{2}$)
<br> From equation (iii) : $\pi_{3} = \pi_{4}$ 
<br> Again $\sum_{i = 1}^{4} \pi_{i} = 1  $
<br> So, $\fbox{$\pi_{1} = \pi_{2} = \pi_{3} = \pi_{4} = 0.25$}$
